### Memory bound in almost everything, compuation is very fast, but data transfer is slow.
Test maximum memory transferrate below. 

```julia
s = 2000
A = CUDA.zeros(Float64, s, s)
B =  CUDA.rand(Float64, s, s)

a = Array(A)
b = Array(B);
```

```julia
@benchmark begin A .= B; synchronize() end
```

```julia
@benchmark begin copyto!($A, $B); synchronize() end
```

### Write you own fast copy kernel on a gpu





![title](l7_cuda_grid.png)

```julia
function copy!(A, B)
    ix = (blockIdx().x-1) * blockDim().x + threadIdx().x
    iy = (blockIdx().y-1) * blockDim().y + threadIdx().y
    @inbounds A[ix,iy] = B[ix,iy]
    return
end

threads = (32, 32)
blocks  = (64, 64)
nx, ny  = threads[1]*blocks[1], threads[2]*blocks[2]
A       = CUDA.zeros(Float64, nx, ny)
B       = CUDA.rand(Float64, nx, ny)

@benchmark begin @cuda blocks=blocks threads=threads copy!(A, B); synchronize() end
```

#### Write a 2d moving average kernel (simplified gaussian blur)

```julia
function moving_average_gpu!(A, B)
    ix = (blockIdx().x-1) * blockDim().x + threadIdx().x
    iy = (blockIdx().y-1) * blockDim().y + threadIdx().y
    
    @inbounds B[ix,iy] = (A[ix+1,iy+1] +  A[ix,iy+1] + A[ix+1,iy] + A[ix,iy]+ A[ix+2,iy+1] + A[ix+1,iy+2] + A[ix+2,iy+2]+ A[ix,iy+2] + A[ix+2,iy])*.1111
    return
end

function moving_average_cpu!(A, B)
    for i in 1:size(B,1)
        for j in 1:size(B,2)
            @inbounds B[i,j] = (A[i+1,j+1] +  A[i,j+1] + A[i+1,j] + A[i,j]+ A[i+2,j+1] + A[i+1,j+2] + A[i+2,j+2]+ A[i,j+2] + A[i+2,j])*.1111
        end
    end
end




function moving_average_cpu_p!(A, B)
    Threads.@threads for i in 1:size(B,1)
        for j in 1:size(B,2)
            @inbounds B[i,j] = (A[i+1,j+1] +  A[i,j+1] + A[i+1,j] + A[i,j]+ A[i+2,j+1] + A[i+1,j+2] + A[i+2,j+2]+ A[i,j+2] + A[i+2,j])*.1111
        end
    end
end
```

### What is actually happening here?

```julia
threads = (32, 32)
blocks  = (4, 4)

nx, ny  = threads[1]*blocks[1], threads[2]*blocks[2]
A       = CUDA.rand(Float64, nx+2, ny+2)
B      = A[2:end-1,2:end-1]

a = Array(A)
b = Array(B);
```

```julia
b[1,1]= 1
fig = Figure(size = (1000,1000))
ax = Axis(fig[1,1])
hm = heatmap!(ax, b, colormap = "thermal")
display(fig)
```

```julia
for i in 1:50
    moving_average_cpu!(a, b)
    @views a[2:end-1,2:end-1] .= b
     if i %1 == 0
        hm = heatmap!(ax, b, colormap = "thermal")
        sleep(0.2)
    end
    
end
```

### Big array on cpu

```julia
threads = (32, 32)
blocks  = (128, 128)

nx, ny  = threads[1]*blocks[1], threads[2]*blocks[2]
A       = CUDA.rand(Float64, nx+2, ny+2)
B      = A[2:end-1,2:end-1]

a = Array(A)
b = Array(B);

b[1,1]= 1
fig = Figure(size = (1000,1000))
ax = Axis(fig[1,1])
hm = heatmap!(ax, b, colormap = "thermal")
display(fig)
```

```julia
for i in 1:300
    moving_average_cpu!(a, b)
    @views a[2:end-1,2:end-1] .= b
     if i %15 == 0
        hm = heatmap!(ax, b, colormap = "thermal")
        sleep(0.0001)
    end
    
end
```

### Now on gpu

```julia
threads = (32, 32)
blocks  = (128, 128)

nx, ny  = threads[1]*blocks[1], threads[2]*blocks[2]
A       = CUDA.rand(Float64, nx+2, ny+2)
B      = A[2:end-1,2:end-1]

B[1,1]= 1
fig = Figure(size = (1000,1000))
ax = Axis(fig[1,1])
hm = heatmap!(ax, Array(B), colormap = "thermal")
display(fig)
```

```julia
for i in 1:300
    @cuda blocks=blocks threads=threads moving_average_gpu!(A, B)
    @views A[2:end-1,2:end-1] .= B
    if i %15 == 0
        hm = heatmap!(ax, Array(B), colormap = "thermal")
        sleep(0.0001)
    end
    
end
```
